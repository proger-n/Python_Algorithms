```python
def sort_check(arr):
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]:
            return True
    return False
    
def find_min(arr):
    min = 1
    for i in range(len(arr)):
        if arr[i] != 0:
            min = arr[i]
            break
    for i in range(len(arr)):
        if arr[i] != 0 and min > arr[i]:
            min = arr[i]
    return min


def TheTaleOfSorting(inp):
    result = []
    for i in range(int(inp[0])):
        # int(input())
        arr = list(map(int, inp[i*2+2].split()))
        n = 0
        while sort_check(arr):
            min = find_min(arr)
            for i in range(len(arr)):
                if arr[i] != 0:
                    arr[i] = arr[i] - min
            n+=min
        result.append(n)
    print(*result, sep='\n')

def test_1():
    inp = ['7',
            '3',
            '1 2 3',
            '5',
            '2 1 2 1 2',
            '4',
            '3 1 5 4',
            '2',
            '7 7',
            '5',
            '4 1 3 2 5',
            '5',
            '2 3 1 4 5',
            '3',
            '1000000000 1 2']
    TheTaleOfSorting(inp)

def main():
    test_1()


if __name__ == "__main__":
    main()
```
